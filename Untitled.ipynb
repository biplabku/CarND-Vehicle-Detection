```python
### Constructing histogram of Oriented Graphs
```

```python
def fetch_hogFeatures(img, orient, pix_per_cell, cell_per_block, vis=False, feature_vec=True):
    if vis == True:
        features, hog_image = hog(img, orientations=orient, 
                                  pixels_per_cell=(pix_per_cell, pix_per_cell),
                                  cells_per_block=(cell_per_block, cell_per_block), 
                                  transform_sqrt=False, 
                                  visualise=vis, feature_vector=feature_vec)
        return features, hog_image
    else:      
        features = hog(img, orientations=orient, 
                       pixels_per_cell=(pix_per_cell, pix_per_cell),
                       cells_per_block=(cell_per_block, cell_per_block), 
                       transform_sqrt=False, 
                       visualise=vis, feature_vector=feature_vec)
        return features

print("Completed")
```

```python
# Applying the hog transform to images
for index in range(10):
    fig, (axs1, axs2) = plt.subplots(1, 2, figsize=(15,11))
    image = cv2.imread(carImages[index])
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    features, hog_image = fetch_hogFeatures(gray, 9, 8, 8, vis=True, feature_vec =True)
    axs1.set_title("Original Image")
    axs1.imshow(image)
    axs2.set_title("Hog Image")
    axs2.imshow(hog_image, cmap ='gray')
```
